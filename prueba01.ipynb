```python
import random

def juego_adivinanza():
    print("ğŸ¯ Â¡BIENVENIDO AL JUEGO DE ADIVINANZA! ğŸ¯")
    print("=" * 40)
    print("Voy a pensar un nÃºmero entre 1 y 100")
    print("Â¡Tienes que adivinarlo!")

    # El nÃºmero secreto
    numero_secreto = random.randint(1, 100)
    intentos = 0
    max_intentos = 7

    while intentos < max_intentos:
        try:
            # Pedir al usuario que adivine
            print(f"\nIntentos restantes: {max_intentos - intentos}")
            guess = int(input("Â¿CuÃ¡l es tu nÃºmero? "))
            intentos += 1

            # Verificar la respuesta
            if guess == numero_secreto:
                print(f"ğŸ‰ Â¡FELICITACIONES! Â¡Adivinaste en {intentos} intentos!")
                print("ğŸ† Â¡Eres increÃ­ble!")
                break
            elif guess < numero_secreto:
                print("ğŸ“ˆ Muy bajo... Â¡Intenta con un nÃºmero mÃ¡s GRANDE!")
            else:
                print("ğŸ“‰ Muy alto... Â¡Intenta con un nÃºmero mÃ¡s PEQUEÃ‘O!")

            # Dar pistas adicionales
            diferencia = abs(guess - numero_secreto)
            if diferencia <= 5:
                print("ğŸ”¥ Â¡CALIENTE! EstÃ¡s muy cerca")
            elif diferencia <= 15:
                print("ğŸŒ¡ï¸ Tibio... te estÃ¡s acercando")
            else:
                print("ğŸ§Š FrÃ­o... estÃ¡s lejos")

        except ValueError:
            print("âŒ Por favor ingresa solo nÃºmeros")
            continue

    # Si se acabaron los intentos
    if intentos >= max_intentos:
        print(f"\nğŸ’” Â¡Se acabaron los intentos!")
        print(f"ğŸ”¢ El nÃºmero era: {numero_secreto}")
        print("ğŸ”„ Â¡Mejor suerte la prÃ³xima vez!")

def estadisticas_juego():
    """FunciÃ³n bonus para llevar estadÃ­sticas"""
    partidas_jugadas = 0
    partidas_ganadas = 0

    while True:
        juego_adivinanza()
        partidas_jugadas += 1

        # Preguntar si ganÃ³ (simplificado)
        resultado = input("\nÂ¿Ganaste esta partida? (s/n): ").lower()
        if resultado == 's':
            partidas_ganadas += 1

        # Mostrar estadÃ­sticas
        porcentaje = (partidas_ganadas / partidas_jugadas) * 100
        print(f"\nğŸ“Š ESTADÃSTICAS:")
        print(f"ğŸ® Partidas jugadas: {partidas_jugadas}")
        print(f"ğŸ† Partidas ganadas: {partidas_ganadas}")
        print(f"ğŸ“ˆ Porcentaje de Ã©xito: {porcentaje:.1f}%")

        # Preguntar si quiere jugar otra vez
        otra_vez = input("\nğŸ”„ Â¿Quieres jugar otra vez? (s/n): ").lower()
        if otra_vez != 's':
            print("\nğŸŒŸ Â¡Gracias por jugar!")
            print(f"ğŸ¯ Tu puntuaciÃ³n final: {partidas_ganadas}/{partidas_jugadas}")
            break

if __name__ == "__main__":
    estadisticas_juego()
```

```python
# Calculadora de Notas para Estudiantes

def agregar_estudiante(estudiantes):
    """Agrega un nuevo estudiante"""
    nombre = input("Nombre del estudiante: ").strip()
    if nombre:
        estudiantes[nombre] = []
        print(f"âœ… Estudiante '{nombre}' agregado exitosamente!")
    else:
        print("âŒ El nombre no puede estar vacÃ­o")

def agregar_nota(estudiantes):
    """Agrega una nota a un estudiante"""
    if not estudiantes:
        print("âŒ No hay estudiantes registrados")
        return

    print("\nEstudiantes disponibles:")
    for i, nombre in enumerate(estudiantes.keys(), 1):
        print(f"{i}. {nombre}")

    try:
        opcion = int(input("\nSelecciona un estudiante (nÃºmero): ")) - 1
        nombres = list(estudiantes.keys())

        if 0 <= opcion < len(nombres):
            estudiante = nombres[opcion]
            materia = input("Materia: ").strip()
            nota = float(input("Nota (0-10): "))

            if 0 <= nota <= 10:
                estudiantes[estudiante].append({
                    'materia': materia,
                    'nota': nota
                })
                print(f"âœ… Nota agregada para {estudiante}!")
            else:
                print("âŒ La nota debe estar entre 0 y 10")
        else:
            print("âŒ OpciÃ³n invÃ¡lida")
    except (ValueError, IndexError):
        print("âŒ Entrada invÃ¡lida")

def calcular_promedio(notas):
    """Calcula el promedio de las notas"""
    if not notas:
        return 0
    return sum(nota['nota'] for nota in notas) / len(notas)

def mostrar_reporte(estudiantes):
    """Muestra el reporte completo"""
    if not estudiantes:
        print("âŒ No hay estudiantes registrados")
        return

    print("\n" + "="*50)
    print("ğŸ“Š REPORTE DE CALIFICACIONES")
    print("="*50)

    for nombre, notas in estudiantes.items():
        print(f"\nğŸ‘¤ ESTUDIANTE: {nombre}")
        print("-" * 30)

        if notas:
            print("ğŸ“š MATERIAS Y NOTAS:")
            for nota_info in notas:
                print(f"   â€¢ {nota_info['materia']}: {nota_info['nota']:.1f}")

            promedio = calcular_promedio(notas)
            print(f"\nğŸ“Š PROMEDIO: {promedio:.2f}")

            # Determinar estado acadÃ©mico
            if promedio >= 9:
                estado = "ğŸ† EXCELENTE"
            elif promedio >= 7:
                estado = "âœ… APROBADO"
            elif promedio >= 6:
                estado = "âš ï¸ REGULAR"
            else:
                estado = "âŒ REPROBADO"

            print(f"ğŸ“ˆ ESTADO: {estado}")

            # EstadÃ­sticas adicionales
            notas_numeros = [n['nota'] for n in notas]
            print(f"ğŸ“Š NOTA MÃS ALTA: {max(notas_numeros):.1f}")
            print(f"ğŸ“Š NOTA MÃS BAJA: {min(notas_numeros):.1f}")
            print(f"ğŸ“š TOTAL MATERIAS: {len(notas)}")
        else:
            print("   Sin notas registradas")

def buscar_estudiante(estudiantes):
    """Busca y muestra informaciÃ³n de un estudiante especÃ­fico"""
    if not estudiantes:
        print("âŒ No hay estudiantes registrados")
        return

    nombre = input("Nombre del estudiante a buscar: ").strip()

    # Buscar coincidencias (no exactas)
    encontrados = []
    for estudiante in estudiantes.keys():
        if nombre.lower() in estudiante.lower():
            encontrados.append(estudiante)

    if encontrados:
        print(f"\nğŸ” RESULTADOS DE BÃšSQUEDA:")
        for estudiante in encontrados:
            notas = estudiantes[estudiante]
            promedio = calcular_promedio(notas) if notas else 0
            print(f"ğŸ‘¤ {estudiante} - Promedio: {promedio:.2f} - Materias: {len(notas)}")
    else:
        print("âŒ No se encontraron estudiantes con ese nombre")

def menu():
    """MenÃº principal"""
    estudiantes = {}

    while True:
        print("\n" + "="*40)
        print("ğŸ“ CALCULADORA DE NOTAS ESTUDIANTILES")
        print("="*40)
        print("1. ğŸ‘¤ Agregar estudiante")
        print("2. ğŸ“ Agregar nota")
        print("3. ğŸ“Š Ver reporte completo")
        print("4. ğŸ” Buscar estudiante")
        print("5. ğŸ“ˆ EstadÃ­sticas generales")
        print("6. ğŸšª Salir")

        try:
            opcion = int(input("\nSelecciona una opciÃ³n (1-6): "))

            if opcion == 1:
                agregar_estudiante(estudiantes)
            elif opcion == 2:
                agregar_nota(estudiantes)
            elif opcion == 3:
                mostrar_reporte(estudiantes)
            elif opcion == 4:
                buscar_estudiante(estudiantes)
            elif opcion == 5:
                mostrar_estadisticas_generales(estudiantes)
            elif opcion == 6:
                print("ğŸ‘‹ Â¡Hasta luego! Que tengas un buen dÃ­a estudiando.")
                break
            else:
                print("âŒ OpciÃ³n invÃ¡lida. Intenta de nuevo.")

        except ValueError:
            print("âŒ Por favor ingresa solo nÃºmeros")

def mostrar_estadisticas_generales(estudiantes):
    """Muestra estadÃ­sticas generales del grupo"""
    if not estudiantes:
        print("âŒ No hay estudiantes registrados")
        return

    total_estudiantes = len(estudiantes)
    estudiantes_con_notas = sum(1 for notas in estudiantes.values() if notas)

    if estudiantes_con_notas == 0:
        print("âŒ No hay notas registradas")
        return

    # Calcular promedios
    promedios = []
    for notas in estudiantes.values():
        if notas:
            promedio = calcular_promedio(notas)
            promedios.append(promedio)

    promedio_general = sum(promedios) / len(promedios)

    print("\n" + "="*40)
    print("ğŸ“Š ESTADÃSTICAS GENERALES")
    print("="*40)
    print(f"ğŸ‘¥ Total de estudiantes: {total_estudiantes}")
    print(f"ğŸ“š Estudiantes con notas: {estudiantes_con_notas}")
    print(f"ğŸ“Š Promedio general del grupo: {promedio_general:.2f}")
    print(f"ğŸ† Promedio mÃ¡s alto: {max(promedios):.2f}")
    print(f"ğŸ“‰ Promedio mÃ¡s bajo: {min(promedios):.2f}")

    # Clasificar estudiantes
    excelentes = sum(1 for p in promedios if p >= 9)
    aprobados = sum(1 for p in promedios if 7 <= p < 9)
    regulares = sum(1 for p in promedios if 6 <= p < 7)
    reprobados = sum(1 for p in promedios if p < 6)

    print(f"\nğŸ“ˆ DISTRIBUCIÃ“N:")
    print(f"ğŸ† Excelentes (9-10): {excelentes}")
    print(f"âœ… Aprobados (7-8.9): {aprobados}")
    print(f"âš ï¸ Regulares (6-6.9): {regulares}")
    print(f"âŒ Reprobados (0-5.9): {reprobados}")

if __name__ == "__main__":
    menu()
```

# Emoji Classifier con PyTorch


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Frases simples con emociones
examples = [
    ("Estoy muy feliz", "ğŸ˜Š"),
    ("Esto es terrible", "ğŸ˜¢"),
    ("Â¡QuÃ© sorpresa!", "ğŸ˜²"),
    ("No me gusta", "ğŸ˜ "),
    ("Estoy muy triste", "ğŸ˜¢"),
    ("Esto es genial", "ğŸ˜Š"),
    ("Estoy enojado", "ğŸ˜ "),
    ("Â¡IncreÃ­ble!", "ğŸ˜²"),
    ("No puedo creerlo", "ğŸ˜²"),
    ("Estoy llorando", "ğŸ˜¢")
]

texts, labels = zip(*examples)

# Convertimos texto a vectores
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts).toarray()

emoji_to_idx = {"ğŸ˜Š": 0, "ğŸ˜¢": 1, "ğŸ˜²": 2, "ğŸ˜ ": 3}
idx_to_emoji = {v: k for k, v in emoji_to_idx.items()}
y = [emoji_to_idx[label] for label in labels]

# DivisiÃ³n entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train = torch.tensor(X_train).float()
y_train = torch.tensor(y_train).long()
X_test = torch.tensor(X_test).float()
y_test = torch.tensor(y_test).long()
```

```python
class EmojiClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(EmojiClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# Crear modelo
input_dim = X_train.shape[1]
model = EmojiClassifier(input_dim=input_dim, hidden_dim=10, output_dim=4)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(100):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    predictions = torch.argmax(model(X_test), dim=1)
    accuracy = (predictions == y_test).float().mean()
    print(f"\nğŸ¯ Test Accuracy: {accuracy:.2f}")
```

```python
def predecir_frase(frase):
    vec = vectorizer.transform([frase]).toarray()
    input_tensor = torch.tensor(vec).float()
    with torch.no_grad():
        output = model(input_tensor)
        pred = torch.argmax(output, dim=1).item()
    return idx_to_emoji[pred]

# Ejemplos para probar
print(predecir_frase("Estoy muy contento"))
print(predecir_frase("QuÃ© horror"))
print(predecir_frase("No me gusta esto"))
print(predecir_frase("Wow, increÃ­ble"))
```
