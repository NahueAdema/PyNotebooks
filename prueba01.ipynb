# **Funciones basicas con python**

```python
import random

def juego_adivinanza():
    print("🎯 ¡BIENVENIDO AL JUEGO DE ADIVINANZA! 🎯")
    print("=" * 40)
    print("Voy a pensar un número entre 1 y 100")
    print("¡Tienes que adivinarlo!")

    # El número secreto
    numero_secreto = random.randint(1, 100)
    intentos = 0
    max_intentos = 7

    while intentos < max_intentos:
        try:
            # Pedir al usuario que adivine
            print(f"\nIntentos restantes: {max_intentos - intentos}")
            guess = int(input("¿Cuál es tu número? "))
            intentos += 1

            # Verificar la respuesta
            if guess == numero_secreto:
                print(f"🎉 ¡FELICITACIONES! ¡Adivinaste en {intentos} intentos!")
                print("🏆 ¡Eres increíble!")
                break
            elif guess < numero_secreto:
                print("📈 Muy bajo... ¡Intenta con un número más GRANDE!")
            else:
                print("📉 Muy alto... ¡Intenta con un número más PEQUEÑO!")

            # Dar pistas adicionales
            diferencia = abs(guess - numero_secreto)
            if diferencia <= 5:
                print("🔥 ¡CALIENTE! Estás muy cerca")
            elif diferencia <= 15:
                print("🌡️ Tibio... te estás acercando")
            else:
                print("🧊 Frío... estás lejos")

        except ValueError:
            print("❌ Por favor ingresa solo números")
            continue

    # Si se acabaron los intentos
    if intentos >= max_intentos:
        print(f"\n💔 ¡Se acabaron los intentos!")
        print(f"🔢 El número era: {numero_secreto}")
        print("🔄 ¡Mejor suerte la próxima vez!")

def estadisticas_juego():
    """Función bonus para llevar estadísticas"""
    partidas_jugadas = 0
    partidas_ganadas = 0

    while True:
        juego_adivinanza()
        partidas_jugadas += 1

        # Preguntar si ganó (simplificado)
        resultado = input("\n¿Ganaste esta partida? (s/n): ").lower()
        if resultado == 's':
            partidas_ganadas += 1

        # Mostrar estadísticas
        porcentaje = (partidas_ganadas / partidas_jugadas) * 100
        print(f"\n📊 ESTADÍSTICAS:")
        print(f"🎮 Partidas jugadas: {partidas_jugadas}")
        print(f"🏆 Partidas ganadas: {partidas_ganadas}")
        print(f"📈 Porcentaje de éxito: {porcentaje:.1f}%")

        # Preguntar si quiere jugar otra vez
        otra_vez = input("\n🔄 ¿Quieres jugar otra vez? (s/n): ").lower()
        if otra_vez != 's':
            print("\n🌟 ¡Gracias por jugar!")
            print(f"🎯 Tu puntuación final: {partidas_ganadas}/{partidas_jugadas}")
            break

if __name__ == "__main__":
    estadisticas_juego()
```

```python
# Calculadora de Notas para Estudiantes

def agregar_estudiante(estudiantes):
    """Agrega un nuevo estudiante"""
    nombre = input("Nombre del estudiante: ").strip()
    if nombre:
        estudiantes[nombre] = []
        print(f"✅ Estudiante '{nombre}' agregado exitosamente!")
    else:
        print("❌ El nombre no puede estar vacío")

def agregar_nota(estudiantes):
    """Agrega una nota a un estudiante"""
    if not estudiantes:
        print("❌ No hay estudiantes registrados")
        return

    print("\nEstudiantes disponibles:")
    for i, nombre in enumerate(estudiantes.keys(), 1):
        print(f"{i}. {nombre}")

    try:
        opcion = int(input("\nSelecciona un estudiante (número): ")) - 1
        nombres = list(estudiantes.keys())

        if 0 <= opcion < len(nombres):
            estudiante = nombres[opcion]
            materia = input("Materia: ").strip()
            nota = float(input("Nota (0-10): "))

            if 0 <= nota <= 10:
                estudiantes[estudiante].append({
                    'materia': materia,
                    'nota': nota
                })
                print(f"✅ Nota agregada para {estudiante}!")
            else:
                print("❌ La nota debe estar entre 0 y 10")
        else:
            print("❌ Opción inválida")
    except (ValueError, IndexError):
        print("❌ Entrada inválida")

def calcular_promedio(notas):
    """Calcula el promedio de las notas"""
    if not notas:
        return 0
    return sum(nota['nota'] for nota in notas) / len(notas)

def mostrar_reporte(estudiantes):
    """Muestra el reporte completo"""
    if not estudiantes:
        print("❌ No hay estudiantes registrados")
        return

    print("\n" + "="*50)
    print("📊 REPORTE DE CALIFICACIONES")
    print("="*50)

    for nombre, notas in estudiantes.items():
        print(f"\n👤 ESTUDIANTE: {nombre}")
        print("-" * 30)

        if notas:
            print("📚 MATERIAS Y NOTAS:")
            for nota_info in notas:
                print(f"   • {nota_info['materia']}: {nota_info['nota']:.1f}")

            promedio = calcular_promedio(notas)
            print(f"\n📊 PROMEDIO: {promedio:.2f}")

            # Determinar estado académico
            if promedio >= 9:
                estado = "🏆 EXCELENTE"
            elif promedio >= 7:
                estado = "✅ APROBADO"
            elif promedio >= 6:
                estado = "⚠️ REGULAR"
            else:
                estado = "❌ REPROBADO"

            print(f"📈 ESTADO: {estado}")

            # Estadísticas adicionales
            notas_numeros = [n['nota'] for n in notas]
            print(f"📊 NOTA MÁS ALTA: {max(notas_numeros):.1f}")
            print(f"📊 NOTA MÁS BAJA: {min(notas_numeros):.1f}")
            print(f"📚 TOTAL MATERIAS: {len(notas)}")
        else:
            print("   Sin notas registradas")

def buscar_estudiante(estudiantes):
    """Busca y muestra información de un estudiante específico"""
    if not estudiantes:
        print("❌ No hay estudiantes registrados")
        return

    nombre = input("Nombre del estudiante a buscar: ").strip()

    # Buscar coincidencias (no exactas)
    encontrados = []
    for estudiante in estudiantes.keys():
        if nombre.lower() in estudiante.lower():
            encontrados.append(estudiante)

    if encontrados:
        print(f"\n🔍 RESULTADOS DE BÚSQUEDA:")
        for estudiante in encontrados:
            notas = estudiantes[estudiante]
            promedio = calcular_promedio(notas) if notas else 0
            print(f"👤 {estudiante} - Promedio: {promedio:.2f} - Materias: {len(notas)}")
    else:
        print("❌ No se encontraron estudiantes con ese nombre")

def menu():
    """Menú principal"""
    estudiantes = {}

    while True:
        print("\n" + "="*40)
        print("🎓 CALCULADORA DE NOTAS ESTUDIANTILES")
        print("="*40)
        print("1. 👤 Agregar estudiante")
        print("2. 📝 Agregar nota")
        print("3. 📊 Ver reporte completo")
        print("4. 🔍 Buscar estudiante")
        print("5. 📈 Estadísticas generales")
        print("6. 🚪 Salir")

        try:
            opcion = int(input("\nSelecciona una opción (1-6): "))

            if opcion == 1:
                agregar_estudiante(estudiantes)
            elif opcion == 2:
                agregar_nota(estudiantes)
            elif opcion == 3:
                mostrar_reporte(estudiantes)
            elif opcion == 4:
                buscar_estudiante(estudiantes)
            elif opcion == 5:
                mostrar_estadisticas_generales(estudiantes)
            elif opcion == 6:
                print("👋 ¡Hasta luego! Que tengas un buen día estudiando.")
                break
            else:
                print("❌ Opción inválida. Intenta de nuevo.")

        except ValueError:
            print("❌ Por favor ingresa solo números")

def mostrar_estadisticas_generales(estudiantes):
    """Muestra estadísticas generales del grupo"""
    if not estudiantes:
        print("❌ No hay estudiantes registrados")
        return

    total_estudiantes = len(estudiantes)
    estudiantes_con_notas = sum(1 for notas in estudiantes.values() if notas)

    if estudiantes_con_notas == 0:
        print("❌ No hay notas registradas")
        return

    # Calcular promedios
    promedios = []
    for notas in estudiantes.values():
        if notas:
            promedio = calcular_promedio(notas)
            promedios.append(promedio)

    promedio_general = sum(promedios) / len(promedios)

    print("\n" + "="*40)
    print("📊 ESTADÍSTICAS GENERALES")
    print("="*40)
    print(f"👥 Total de estudiantes: {total_estudiantes}")
    print(f"📚 Estudiantes con notas: {estudiantes_con_notas}")
    print(f"📊 Promedio general del grupo: {promedio_general:.2f}")
    print(f"🏆 Promedio más alto: {max(promedios):.2f}")
    print(f"📉 Promedio más bajo: {min(promedios):.2f}")

    # Clasificar estudiantes
    excelentes = sum(1 for p in promedios if p >= 9)
    aprobados = sum(1 for p in promedios if 7 <= p < 9)
    regulares = sum(1 for p in promedios if 6 <= p < 7)
    reprobados = sum(1 for p in promedios if p < 6)

    print(f"\n📈 DISTRIBUCIÓN:")
    print(f"🏆 Excelentes (9-10): {excelentes}")
    print(f"✅ Aprobados (7-8.9): {aprobados}")
    print(f"⚠️ Regulares (6-6.9): {regulares}")
    print(f"❌ Reprobados (0-5.9): {reprobados}")

if __name__ == "__main__":
    menu()
```

```python
import random

# Base de datos de elementos para las historias
personajes = [
    "un valiente caballero", "una hechicera misteriosa", "un robot del futuro",
    "una princesa rebelde", "un pirata aventurero", "un detective astuto",
    "una científica brillante", "un dragón amigable", "un ninja sigiloso",
    "una bailarina mágica", "un cocinero extraordinario", "un explorador intrépido"
]

lugares = [
    "en un castillo encantado", "en una nave espacial", "en una isla desierta",
    "en un laboratorio secreto", "en un circo mágico", "en las montañas nevadas",
    "en una ciudad submarina", "en un bosque misterioso", "en un desierto infinito",
    "en una biblioteca antigua", "en un parque de diversiones", "en una cueva cristalina"
]

objetos_magicos = [
    "una espada brillante", "un libro de hechizos", "una llave dorada",
    "un collar de estrellas", "un mapa del tesoro", "una poción burbujeante",
    "un espejo parlante", "una varita mágica", "una brújula encantada",
    "un anillo invisible", "una capa voladora", "una lámpara misteriosa"
]

problemas = [
    "había perdido su memoria", "necesitaba salvar el mundo",
    "buscaba un tesoro legendario", "tenía que resolver un misterio",
    "quería encontrar el camino a casa", "debía rescatar a un amigo",
    "necesitaba aprender una lección importante", "tenía que enfrentar su mayor miedo",
    "buscaba la respuesta a una pregunta antigua", "quería cumplir un sueño imposible",
    "debía proteger algo muy valioso", "necesitaba demostrar su valor"
]

soluciones = [
    "usando su inteligencia y astucia", "con la ayuda de nuevos amigos",
    "descubriendo un poder oculto en su interior", "siguiendo las pistas correctas",
    "siendo valiente en el momento crucial", "aprendiendo a trabajar en equipo",
    "confiando en sus instintos", "usando el objeto mágico sabiamente",
    "recordando las palabras de un sabio", "nunca rindiéndose ante las dificultades",
    "encontrando la magia en las cosas simples", "siendo fiel a sus principios"
]

finales = [
    "y vivieron felices para siempre", "y se convirtió en una leyenda",
    "y encontró la paz que buscaba", "y descubrió el verdadero significado de la amistad",
    "y regresó a casa como un héroe", "y su historia inspiró a muchos otros",
    "y aprendió que el verdadero tesoro estaba en el viaje", "y encontró su lugar en el mundo",
    "y cambió para siempre su destino", "y demostró que los sueños se pueden cumplir",
    "y su valor fue recompensado", "y escribió una nueva página en la historia"
]

def generar_historia_simple():
    """Genera una historia aleatoria básica"""
    personaje = random.choice(personajes)
    lugar = random.choice(lugares)
    objeto = random.choice(objetos_magicos)
    problema = random.choice(problemas)
    solucion = random.choice(soluciones)
    final = random.choice(finales)

    historia = f"""
🌟 ¡HISTORIA GENERADA! 🌟
========================

Había una vez {personaje} que vivía {lugar}.
Un día, mientras exploraba, encontró {objeto}.

Sin embargo, nuestro héroe {problema} y necesitaba ayuda urgentemente.

Afortunadamente, logró superar el desafío {solucion},
y al final {final}.

========================
    """

    return historia

def generar_historia_personalizada():
    """Permite al usuario personalizar algunos elementos"""
    print("🎨 ¡Vamos a crear una historia personalizada!")
    print("Puedes elegir algunos elementos o dejarlos aleatorios")
    print()

    # Elegir personaje
    print("👤 PERSONAJES DISPONIBLES:")
    for i, p in enumerate(personajes[:6], 1):  # Mostrar solo algunos
        print(f"{i}. {p}")
    print("7. 🎲 Sorpréndeme (aleatorio)")

    try:
        opcion = int(input("\nElige tu personaje (1-7): "))
        if 1 <= opcion <= 6:
            personaje_elegido = personajes[opcion-1]
        else:
            personaje_elegido = random.choice(personajes)
    except ValueError:
        personaje_elegido = random.choice(personajes)

    # Elegir lugar
    print("\n🗺️ LUGARES DISPONIBLES:")
    for i, l in enumerate(lugares[:6], 1):
        print(f"{i}. {l}")
    print("7. 🎲 Sorpréndeme (aleatorio)")

    try:
        opcion = int(input("\nElige el lugar (1-7): "))
        if 1 <= opcion <= 6:
            lugar_elegido = lugares[opcion-1]
        else:
            lugar_elegido = random.choice(lugares)
    except ValueError:
        lugar_elegido = random.choice(lugares)

    # Generar el resto aleatoriamente
    objeto = random.choice(objetos_magicos)
    problema = random.choice(problemas)
    solucion = random.choice(soluciones)
    final = random.choice(finales)

    historia = f"""
🌟 ¡TU HISTORIA PERSONALIZADA! 🌟
================================

Había una vez {personaje_elegido} que vivía {lugar_elegido}.
Un día, mientras exploraba, encontró {objeto}.

Sin embargo, nuestro héroe {problema} y necesitaba ayuda urgentemente.

Afortunadamente, logró superar el desafío {solucion},
y al final {final}.

================================
    """

    return historia

def generar_historia_loca():
    """Genera una historia completamente disparatada"""
    # Elementos más locos y divertidos
    personajes_locos = [
        "un pinguino que habla", "una pizza viviente", "un calcetín perdido",
        "un superhéroe que le tiene miedo a los ratones", "una computadora enamorada",
        "un cactus bailarín"
    ]

    lugares_locos = [
        "en una lavadora gigante", "en el interior de un sandwich",
        "en un planeta hecho de queso", "en la nevera de tu abuela",
        "en un mundo donde todo está al revés", "en una nube de algodón de azúcar"
    ]

    objetos_locos = [
        "un tenedor cantante", "unos zapatos que vuelan hacia atrás",
        "una calculadora que solo suma chistes", "un paraguas que llueve hacia arriba",
        "un reloj que va al pasado", "un lápiz que dibuja el futuro"
    ]

    personaje = random.choice(personajes_locos)
    lugar = random.choice(lugares_locos)
    objeto = random.choice(objetos_locos)
    problema = random.choice(problemas)
    solucion = random.choice(soluciones)
    final = random.choice(finales)

    historia = f"""
🤪 ¡HISTORIA SÚPER LOCA! 🤪
==========================

En un mundo muy extraño, {personaje} vivía {lugar}.
Un martes cualquiera, encontró {objeto} debajo de su cama.

Pero resulta que {problema}, lo cual era bastante inconveniente.

Por suerte, resolvió todo {solucion}, porque así son las cosas en este mundo raro.

Al final {final}, y todos se rieron mucho.

¡FIN! 😂
==========================
    """

    return historia

def guardar_historia(historia):
    """Guarda la historia en un archivo"""
    try:
        nombre_archivo = input("¿Nombre para tu archivo? (sin extensión): ").strip()
        if not nombre_archivo:
            nombre_archivo = f"historia_{random.randint(1000, 9999)}"

        with open(f"{nombre_archivo}.txt", "w", encoding="utf-8") as archivo:
            archivo.write(historia)

        print(f"✅ Historia guardada como '{nombre_archivo}.txt'")
    except Exception as e:
        print(f"❌ Error al guardar: {e}")

def menu():
    """Menú principal del generador"""
    print("📚 ¡BIENVENIDO AL GENERADOR DE HISTORIAS! 📚")
    print("=" * 45)

    historias_generadas = 0

    while True:
        print(f"\n📊 Historias generadas en esta sesión: {historias_generadas}")
        print("\n🎭 ¿Qué tipo de historia quieres?")
        print("1. 🎲 Historia completamente aleatoria")
        print("2. 🎨 Historia personalizada (tú eliges algunos elementos)")
        print("3. 🤪 Historia súper loca y divertida")
        print("4. 📜 Ver elementos disponibles")
        print("5. 🚪 Salir")

        try:
            opcion = int(input("\nElige una opción (1-5): "))

            if opcion == 1:
                historia = generar_historia_simple()
                print(historia)
                historias_generadas += 1

            elif opcion == 2:
                historia = generar_historia_personalizada()
                print(historia)
                historias_generadas += 1

            elif opcion == 3:
                historia = generar_historia_loca()
                print(historia)
                historias_generadas += 1

            elif opcion == 4:
                mostrar_elementos()

            elif opcion == 5:
                print(f"\n👋 ¡Hasta luego! Generaste {historias_generadas} historias.")
                print("🌟 ¡Sigue creando y soñando!")
                break

            else:
                print("❌ Opción inválida")
                continue

            # Preguntar si quiere guardar (solo después de generar una historia)
            if opcion in [1, 2, 3]:
                guardar = input("\n💾 ¿Quieres guardar esta historia? (s/n): ").lower()
                if guardar == 's':
                    guardar_historia(historia)

        except ValueError:
            print("❌ Por favor ingresa solo números")

def mostrar_elementos():
    """Muestra todos los elementos disponibles"""
    print("\n" + "="*50)
    print("📋 ELEMENTOS DISPONIBLES PARA LAS HISTORIAS")
    print("="*50)

    print(f"\n👤 PERSONAJES ({len(personajes)} disponibles):")
    for p in personajes:
        print(f"   • {p}")

    print(f"\n🗺️ LUGARES ({len(lugares)} disponibles):")
    for l in lugares:
        print(f"   • {l}")

    print(f"\n✨ OBJETOS MÁGICOS ({len(objetos_magicos)} disponibles):")
    for o in objetos_magicos:
        print(f"   • {o}")

    print(f"\n🎯 Total de combinaciones posibles: {len(personajes) * len(lugares) * len(objetos_magicos) * len(problemas) * len(soluciones) * len(finales):,}")

if __name__ == "__main__":
    menu()
```

# Emoji Classifier con PyTorch


```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Frases simples con emociones
examples = [
    ("Estoy muy feliz", "😊"),
    ("Esto es terrible", "😢"),
    ("¡Qué sorpresa!", "😲"),
    ("No me gusta", "😠"),
    ("Estoy muy triste", "😢"),
    ("Esto es genial", "😊"),
    ("Estoy enojado", "😠"),
    ("¡Increíble!", "😲"),
    ("No puedo creerlo", "😲"),
    ("Estoy llorando", "😢")
]

texts, labels = zip(*examples)

# Convertimos texto a vectores
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts).toarray()

emoji_to_idx = {"😊": 0, "😢": 1, "😲": 2, "😠": 3}
idx_to_emoji = {v: k for k, v in emoji_to_idx.items()}
y = [emoji_to_idx[label] for label in labels]

# División entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train = torch.tensor(X_train).float()
y_train = torch.tensor(y_train).long()
X_test = torch.tensor(X_test).float()
y_test = torch.tensor(y_test).long()
```

```python
class EmojiClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(EmojiClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# Crear modelo
input_dim = X_train.shape[1]
model = EmojiClassifier(input_dim=input_dim, hidden_dim=10, output_dim=4)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(100):
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    predictions = torch.argmax(model(X_test), dim=1)
    accuracy = (predictions == y_test).float().mean()
    print(f"\n🎯 Test Accuracy: {accuracy:.2f}")
```

```python
def predecir_frase(frase):
    vec = vectorizer.transform([frase]).toarray()
    input_tensor = torch.tensor(vec).float()
    with torch.no_grad():
        output = model(input_tensor)
        pred = torch.argmax(output, dim=1).item()
    return idx_to_emoji[pred]

# Ejemplos para probar
print(predecir_frase("Estoy muy contento"))
print(predecir_frase("Qué horror"))
print(predecir_frase("No me gusta esto"))
print(predecir_frase("Wow, increíble"))
```
